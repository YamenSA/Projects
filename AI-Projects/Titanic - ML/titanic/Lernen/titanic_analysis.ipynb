```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
print(df.shape)
```

```python
print(df.columns)
```

```python
print(df['Age'])
```

```python
max_alter = df['Age'].max()
print(max_alter)
```

```python
erste_klasse_passagiere = df[df['Pclass']==1]
print(len(erste_klasse_passagiere))
```

```python
überlebensrate = df[df['Sex']=='female']
print(überlebensrate['Survived'].mean())
```

```python
anzahl_leer = df['Age'].isna().sum()
print(f"Es gibt {anzahl_leer} leere Felder in dieser Spalte.")
```

```python
age_median = df['Age'].median()
print(f"Der Median des Alters ist: {age_median}")
df['Age'] = df['Age'].fillna(age_median)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(df['Age'], bins= 30, kde=True)
```

```python
# 1. Daten vorbereiten (hast du schon mit df)

# 2. Die Grafik erstellen (mit Seaborn)
sns.countplot(data=df, x='Sex', hue='Survived')

# 3. Das Layout anpassen (mit Matplotlib)
plt.title('Überlebte nach Geschlecht')
plt.xlabel('Geschlecht')
plt.ylabel('Anzahl Überlebender')
# 4. Anzeigen
plt.show()
```
